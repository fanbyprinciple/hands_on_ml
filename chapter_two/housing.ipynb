```python
# small function to fetch data
import os
import tarfile
from six.moves import urllib


DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = "datasets/housing"
HOUSING_URL = DOWNLOAD_ROOT + HOUSING_PATH + "/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(hosuing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

```python
# create the test set
# just pick some instances randomly 20% of the dataset
# keep them aside
import numpy as np

# dummy = (np.random.permutation(20))
# print(dummy)



def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
#     print(train_indices)
#     print(data.iloc[train_indices[:5]])
    return data.iloc[train_indices], data.iloc[test_indices]
    
```

```python
train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set), len(test_set))

# since this will always create a new dataseteverytime program is run
# one way to fix permutation is through a random.seed(42)
```

```python
# An alternate way is to make a hash to create a test set, this 
# ensures that the hash will remain consistent across multiple refresh of dataset
# a sample implementation
# if this value is lower or equal to 256

import hashlib

hh = hashlib.md5(np.int64(64))
print(type(hh))
print(hh.digest())

# for i in hh.digest():
#     print(i)
print(hh.digest()[-1])


def test_set_check(identifier, test_ratio, hash):
    return hash (np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id : test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# in order to reduce the nnumber of strata or division in
# average income it is important that we reduce the no. of category of average income
# we do it by

housing["income_cat"] = np.ceil(housing["median_income"] / 1.5)
print(housing["income_cat"][:5]).0
# replace greater than 5 with 5
housing["income_cat"].where(housing["income_cat"] < 5,5.0, inplace=True)
print(housing["income_cat"][:5])
```

```python
# we will do stratified sampling using income category
# so that we get the no.of test and train from all the income catefory equally

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index,test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
```
